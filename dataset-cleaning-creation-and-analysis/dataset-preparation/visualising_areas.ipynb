## Visualising the coverage area


```python
import numpy as np
import pandas as pd
import math
import time
import matplotlib.pyplot as plt
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
tmc_oth_df = pd.read_csv("csv_files/tmc_oth.csv")
```

```python
polygons_tmc = []
for i in range(855):
    polygons_tmc.append(Polygon([(tmc_oth_df['ul_la'][i], tmc_oth_df['ul_lo'][i]), (tmc_oth_df['ur_la'][i], tmc_oth_df['ur_lo'][i]), (tmc_oth_df['lr_la'][i], tmc_oth_df['lr_lo'][i]), (tmc_oth_df['ll_la'][i], tmc_oth_df['ll_lo'][i])]))
```

```python
for i in range(855):
    plt.plot(*polygons_tmc[i].exterior.xy)
    plt.xlim(-60, 60)
```
